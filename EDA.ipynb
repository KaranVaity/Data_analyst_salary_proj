```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('salary_data_cleaned.csv')
df
```

```python
#Show all the columns
df.columns
```

```python
#Plot a bar plot to show the number of job postings in each state
title_dict = {'fontsize': 30,
 'fontweight' : 'bold',}
label_dict = {'fontsize': 15,
 'fontweight' : 'normal',
  }
stitle_dict = {'fontsize': 20,
 'fontweight' : 'bold',}

plt.figure(figsize=(15,9))
ax = sns.countplot(x = 'State', data = df)
ax.set_title('The Number of Job Postings in each State', fontdict = title_dict);
ax.set_xlabel('States', fontdict = label_dict);
ax.set_ylabel('Number of job postings',fontdict=label_dict);
```

```python
#Plot a bar plot to show the salary estimates
plt.figure(figsize=(15,9))
ax = sns.distplot(df['Avg_salary'])
ax.set_title('Salary distribution', fontdict = title_dict);
ax.set_xlabel('Average Salary', fontdict = label_dict);
ax.set_ylabel('Number of job postings',fontdict=label_dict);
```

```python
#Countplot to show required skills in the job postings
skills = ['python', 'r', 'tableau', 'sas', 'spark', 'excel', 'splunk', 'sql']
f = plt.figure(figsize = (20,20));


for x, y in enumerate(skills,start=1):
    ax = plt.subplot(4,2,x);
    plt.tight_layout(pad = 10);
    sns.countplot(x = y, data = df, ax = ax);
    ax.set_title('Number of job postings requiring '+y, fontdict = stitle_dict);
    ax.set_ylabel('Number of job postings', fontdict = label_dict);
    ax.set_xlabel(y,fontdict=label_dict);

f.suptitle('Job postings based on skills', size = 50, color = 'purple', weight = 'bold');
```

```python
#Job postings and number of the competitors

plt.figure(figsize=(15,9))
ax = sns.countplot(x = 'Num_of_comp', data = df)
ax.set_title('The Number of Job Postings', fontdict = title_dict);
ax.set_xlabel('Number of competitors', fontdict = label_dict);
ax.set_ylabel('Number of job postings',fontdict=label_dict);
```

```python
#Job postings and size of employees of the company

plt.figure(figsize=(15,9))
ax = sns.countplot(y = 'Size', data = df)
ax.set_title('The Number of Job Postings', fontdict = title_dict);
ax.set_xlabel('Size of company', fontdict = label_dict);
ax.set_ylabel('Number of job postings',fontdict=label_dict);
```

```python
#Job postings and type of ownership

plt.figure(figsize=(15,9))
ax = sns.countplot(y = 'Type of ownership', data = df)
ax.set_title('Job postings based on Ownership of the company', fontdict = title_dict);
ax.set_xlabel('Type of ownership', fontdict = label_dict);
ax.set_ylabel('Number of job postings',fontdict=label_dict);
```

```python
#Job postings and age of the company

plt.figure(figsize=(15,9))
ax = sns.distplot(df['Age'])
ax.set_title('Job postings based on the Age', fontdict = title_dict);
ax.set_xlabel('Age', fontdict = label_dict);
ax.set_ylabel('Number of job postings',fontdict=label_dict);
```

```python
#Job postings and Rating of the company
plt.figure(figsize=(15,9))
ax = sns.distplot(df['Rating'])
ax.set_title('Job postings based on the Rating', fontdict = title_dict);
ax.set_xlabel('Rating', fontdict = label_dict);
ax.set_ylabel('Number of job postings',fontdict=label_dict);
```
